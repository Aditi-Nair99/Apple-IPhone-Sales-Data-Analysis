# TOP 10 HIGHEST RATED IPHONES ON FLIPKART IN INDIA.

```python
data.head()
```

```python
highest_rated = data.sort_values(by = ["Star Rating"], ascending = False)
highest_rated = highest_rated.head(10)
```

```python
print(highest_rated['Product Name'])
```

```python
print(highest_rated)
```

THE TOP 10 HIGHEST RATED APPLE IPHONES ARE DISPLAYED.

# HOW MANY RATINGS DO THE HIGHEST RATED IPHONES ON FLIPKART HAVE ?

```python
data.head()
```

```python
iphones = highest_rated['Product Name'].value_counts()
label = iphones.index
counts = highest_rated["Number Of Ratings"]
figure = px.bar(highest_rated, x = label, y = counts, title = "Numer of Rattings of Highest Rated IPhones")
figure.show()
```

# WHICH IPHONE HAS THE HIGHEST NUMBER OF REVIEWS ON FLIPKART ?

```python
data.head()
```

```python
iphones = highest_rated['Product Name'].value_counts()
label = iphones.index
counts = highest_rated["Number Of Reviews"]
figure = px.bar(highest_rated, x = label, y = counts, title = "Number of Reviews of Highest Rated IPhones")
figure.show()
```

# WHAT IS THE RELATIONSHIP BETWEEN THE SALE PRICES OF IPHONES AND THE NUMBER OF RATINGS ON FLIPKART ? 

```python
data.head()
```

```python
import plotly.express as px

figure = px.scatter(
    data_frame=data,
    x="Number Of Ratings",
    y="Sale Price",
    size="Discount Percentage",
    trendline="ols",
    title="Relationship between Sale Price and Number of Ratings"
)

figure.show()
```

THERE IS A NEGATIVE RELATIONSHIP BETWEEN SALE PRICE AND NUMBER OF RATING. IT MEANS THE IPHONES WITH LOWER SALES PRICES ARE SOLD MORE IN INDIA. 

# WHAT IS THE RELATIONSHIP BETWEEN THE DISCOUNT PERCENTAGE AND THE NUMBER OF RATINGS OF IPHONES ON FLIPKART ?

```python
data.head()
```

```python
import plotly.express as px

figure = px.scatter(
    data_frame=data,
    x="Number Of Ratings",
    y="Discount Percentage",
    size="Sale Price",
    trendline="ols",
    title="Relationship between Discount Percentage and Number of Ratings"
)

figure.show()
```

# CAN YOU FIGURE OUT THE MOST EXPENSIVE AND LEAST EXPENSIVE IPHONES IN THE INDIAN MARKET ALONG WITH ALL THEIR SPECIFICATIONS ?

```python
data
```

```python
most_expensive = data.loc[data['Sale Price'].idxmax()]
least_expensive = data.loc[data['Sale Price'].idxmin()]

# Display the result
print("Most Expensive Product:")
print(most_expensive)

print("\n Least Expensive Product:")
print(least_expensive)
```
